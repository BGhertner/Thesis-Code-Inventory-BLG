```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as mpatches

plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.usetex'] = True
mpl.rcParams["font.size"] = 15

%matplotlib inline
```

```python
fig, ax = plt.subplots(layout='constrained')
fig.set_size_inches(3.8,3)
fig.set_dpi(250)

xx = np.linspace(-2*np.pi, 2*np.pi)

amps = [1.0, 0.2, 0.01, 0.01]

edge = np.zeros_like(xx)
for j, amp in enumerate(amps):
    edge += amp*np.cos(xx*j)

ax.set(ylim=(0.0, 2.0), xlim=(xx[0], xx[-1]))
ax.plot(xx, edge, '--', color='brown')
ax.fill_between(xx, edge, 0, color='grey', alpha=0.3)
ax.fill_between(xx, 2., edge, color='cyan', alpha=0.3)
ax.set_xticks([])
ax.set_yticks([])
ax.annotate('Clear: Dry Air', (-6.0, 1.8))
ax.annotate('Water Vapour Only', (-4.0, 1.6))
ax.annotate('Cloud: Dry Air', (-6.0, 0.3))
ax.annotate('Water Vapour \& Liquid Water', (-3.9, 0.1))

plt.show()
```

```python
fig, ax = plt.subplots(layout='constrained')
fig.set_size_inches(4.1,2.8)
fig.set_dpi(250)

xx = np.linspace(-2*np.pi, 2*np.pi)

amps = [1.0, 0.2, 0.01, 0.01]

edge = np.zeros_like(xx)
for j, amp in enumerate(amps):
    edge += amp*np.cos(xx*j)

edge_m = np.zeros_like(xx)
for j, amp in enumerate(amps):
    if j==0: edge_m += amp*np.cos(xx*j)
    else: edge_m += amp/2*np.cos(xx*j)

ax.set(ylim=(0.0, 2.0), xlim=(xx[0], xx[-1]))
ax.plot(xx, edge, '--', color='brown')
ax.plot(xx, edge_m, '--', color='rebeccapurple')
ax.fill_between(xx, edge, 0, color='grey', alpha=0.3)
ax.fill_between(xx, 2., edge, color='cyan', alpha=0.3)
ax.set_xticks([])
ax.set_yticks([])
ax.annotate('Clear', (-6.0, 1.8))
ax.annotate('Cloud', (-6.0, 0.1))

arr = mpatches.FancyArrowPatch((-5.6, 1.5), (-5.6, 0.6),
                               arrowstyle='->,head_width=.15', mutation_scale=20)
ax.add_patch(arr)
ax.annotate(r'$\frac{d\bar\ell}{dz} < 0$', (7.0, -0.25), xycoords=arr, 
            ha='center', va='bottom', size=22, color='darkcyan')

ax.annotate(r"$H+\mu Z_c'$", (0.7,0.63), color='brown')
ax.annotate(r"$H+\mu Z_m'$", (1.7,1.05), color='rebeccapurple')


plt.show()
```

```python
fig, ax = plt.subplots(layout='constrained')
fig.set_size_inches(4.1,2.8)
fig.set_dpi(250)

xx = np.linspace(-2*np.pi, 2*np.pi)

amps = [1.0, -0.2, -0.01, -0.01]

edge = np.zeros_like(xx)
for j, amp in enumerate(amps):
    edge += amp*np.cos(xx*j)

edge_m = np.zeros_like(xx)
for j, amp in enumerate(amps):
    if j==0: edge_m += amp*np.cos(xx*j)
    else: edge_m += -amp/2*np.cos(xx*j)

ax.set(ylim=(0.0, 2.0), xlim=(xx[0], xx[-1]))
ax.plot(xx, edge, '--', color='brown')
ax.plot(xx, edge_m, '--', color='rebeccapurple')
ax.fill_between(xx, edge, 0, color='cyan', alpha=0.3)
ax.fill_between(xx, 2., edge, color='grey', alpha=0.3)
ax.set_xticks([])
ax.set_yticks([])
ax.annotate('Cloud', (-6.0, 1.8))
ax.annotate('Clear', (-6.0, 0.1))

arr = mpatches.FancyArrowPatch((-5.6, 0.5), (-5.6, 1.4),
                               arrowstyle='->,head_width=.15', mutation_scale=20)
ax.add_patch(arr)
ax.annotate(r'$\frac{d\bar\ell}{dz} > 0$', (7., 0.8), xycoords=arr, 
            ha='center', va='bottom', size=22, color='darkcyan')

ax.annotate(r"$H+\mu Z_m'$", (-1.3,1.22), color='rebeccapurple')
ax.annotate(r"$H+\mu Z_c'$", (0.,0.6), color='brown')


plt.show()
```
